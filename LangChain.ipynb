```python
from langchain_google_genai import ChatGoogleGenerativeAI
from transformers import pipeline, AutoTokenizer, AutoModelForCausalLM
```

### Set up your Gemini API key
To use the Gemini API, you'll need an API key. If you don't already have one, create a key in Google AI Studio. In Colab, add the key to the secrets manager under the "🔑" icon in the left panel. Give it the name `GOOGLE_API_KEY`. Then, we can securely access it:

```python
# Used to securely store your API key
from google.colab import userdata

GOOGLE_API_KEY = userdata.get('GEMINI_API_KEY_1')
```

### Initialize the Gemini model
Now we can initialize the `ChatGoogleGenerativeAI` model using the API key.

```python
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-2.5-flash", google_api_key=GOOGLE_API_KEY, temperature=0.7, max_output_tokens=500)
```

### Create a Chat Prompt Template
Next, let's create a `ChatPromptTemplate` to define the structure of our prompts.

```python



from langchain_core.prompts import ChatPromptTemplate

prompt = ChatPromptTemplate.from_messages([
    ("system", "You are a helpful AI assistant. Answer all questions to the best of your ability."),
    ("user", "{input}")
])
```

```python

chain = prompt | llm

# Invoke the chain with a sample input
chain.invoke({"input": "how are you?"})
```

```python
tokenizer = AutoTokenizer.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")
model_hf = AutoModelForCausalLM.from_pretrained("TinyLlama/TinyLlama-1.1B-Chat-v1.0")

pipeline_hf = pipeline("text-generation", model=model_hf, tokenizer=tokenizer, max_new_tokens=200)
```

```python
from langchain_huggingface import HuggingFacePipeline

llm_hf = HuggingFacePipeline(pipeline=pipeline_hf)
```

```python
result = llm_hf.invoke("what is good boy")
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
indian_cricketers_sentences = [
    "Virat Kohli is known for his aggressive batting style and consistency across all formats of cricket.",
    "Rohit Sharma holds the record for the highest individual score in One Day Internationals with 264 runs.",
    "MS Dhoni led India to World Cup victories in both 2007 T20 and 2011 ODI tournaments.",
    "Jasprit Bumrah is famous for his lethal yorkers and death-over bowling in limited-overs cricket.",
    "Ravindra Jadeja is a brilliant all-rounder, contributing with both the bat and ball for India."
]
```

```python
query='tell me about Bhumra'
```

```python
ind=model.encode_document(indian_cricketers_sentences)
query=model.encode(query)
```

```python
sore=cosine_similarity([query],ind)[0]
```

```python

index,score=sorted(list(enumerate(sore)),key=lambda x:x[1])[-1]
```

```python
print(indian_cricketers_sentences[index])
```

```python
print("similarity",score)
```
